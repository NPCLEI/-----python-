```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
plt.figure(1,figsize=(20,20))

def histMean(raw):
    h,w = raw.shape
    #计算图片灰度级，从零开始，所以+1
    Dm = np.max(raw) + 1

    dct = [0 for i in range(Dm)]
    #图片像素点总个数
    A0 = h*w
    #Dm/A0
    C = Dm / A0
    raw = raw.reshape((1,A0))

    #计算各个灰度级像素的个数
    for i in range(A0):
        dct[raw[0,i]]+=1

    #计算直方图均衡后的像素对应的灰度
    for i in range(Dm):
        dct[i] = round(C * dct[i])

    #将新的灰度值赋值回图像
    for i in range(A0):
        raw[0,i] = dct[raw[0,i]]

    raw = raw.reshape((h,w))

    return raw

raw=cv2.imread('../imgs/c2/cells.png',0)

plt.subplot(1,4,1)
plt.imshow(raw,cmap='gray')
plt.subplot(1,4,2)
plt.hist(raw)

plt.subplot(1,4,3)
plt.imshow(histMean(raw),cmap='gray')
plt.subplot(1,4,4)
plt.hist(raw)
plt.show()
```
