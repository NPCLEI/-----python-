2.乘法运算，这里的乘法是M1ij * M2ij，相同位置的元素相乘

```python
import utils
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1,figsize=(20,20))

img1 = utils.read('../imgs/c2/muti1.jpg')
img2 = utils.readGray('../imgs/c2/muti2.jpg')
#归一化，不归一化会出现255*255的情况，会溢出
img2 = img2 / np.max(img2)

plt.subplot(1,3,1)
plt.imshow(img1)

plt.subplot(1,3,2)
plt.imshow(img2,cmap='gray')

#三个通道都要乘一遍
for i in range(3):
    img1[:,:,i] = img1[:,:,i] * img2

plt.subplot(1,3,3)
plt.imshow(img1)
```
